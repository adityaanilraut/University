```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/Users/aditya/Desktop/project 2/new.csv'
data = pd.read_csv(file_path)

# Display the first few rows to understand its structure
data.head()
```

```python
# Extracting data from the DataFrame for plotting
sizes = data['size']
bubbleSort_times = data['bubbleSort']
insertionSort_times = data['insertionSort']
selectionSort_times = data['selectionSort']
mergeSort_times = data['mergesort']
quickSort_times = data['quicksort']
methodSort_times = data['methodsort']

# Plotting the data
plt.figure(figsize=(10, 6))

# Plot each sorting algorithm's data
plt.plot(sizes, bubbleSort_times, label="Bubble Sort")
plt.plot(sizes, insertionSort_times, label="Insertion Sort")
plt.plot(sizes, selectionSort_times, label="Selection Sort")
plt.plot(sizes, mergeSort_times, label="Merge Sort")
plt.plot(sizes, quickSort_times, label="Quick Sort")
plt.plot(sizes, methodSort_times, label="Method Sort")

# Adding titles and labels
plt.title("Sorting Algorithms: Time Complexity Comparison")
plt.xlabel("Input Size")
plt.ylabel("Time Taken (seconds)")
plt.yscale('log')  # Log scale for better visualization of time differences

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
