```python
#a. Read in the dataset from the URL to the .csv file above.
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/roualdes/data/master/carnivora.csv')
df.head()
```

```python
#b. Provide a table of the counts of the number of animals in SuperFamily versus Order.
superfamily_order_table = pd.crosstab(df['SuperFamily'], df['Order'])

superfamily_order_table
```

```python
#c. Calculate a new column named long_gestation_length which records True 
#  for all animals with a gestation length (GL) greater or equal to the mean gestation length, and False otherwise.

mean_gl = df['GL'].mean()
df['long_gestation_length'] = df['GL'] >= mean_gl

#d. Change the type of the column long_gestation_length from bool to category.
df['long_gestation_length'] = df['long_gestation_length'].astype('category')

```

```python
#e. Use rename_categories to rename the values of the column long_gestation_length from True/False to "long"/"short".
df['long_gestation_length'] = df['long_gestation_length'].cat.rename_categories({True: 'long', False: 'short'})


df[['GL', 'long_gestation_length']].head()
```

```python
#Provide a table of proportions of SuperFamily by the variable long_gestation_length, where the rows (index) are normalized.
#Which SuperFamily is more likely to have relatively shorter gestation lengths?
# Create a crosstab with proportions of long_gestation_length by SuperFamily
proportions_table = pd.crosstab(df['SuperFamily'], df['long_gestation_length'], normalize='index')
superfamily_most_short = proportions_table.sort_values(by='short', ascending=False)
# Display the table of proportions
proportions_table
```

```python
#g. Calculate a new column that records the average brain weight (SB) to average body weight (SW) ratio for all animals in the DataFrame.

df['brain_body_ratio'] = df['SB'] / df['SW']

family_brain_body_ratio = df.groupby('Family')['brain_body_ratio'].mean()
family_brain_body_ratio
```

```python
#h. Provide a list of the top three Family ranked by mean brain weight (SB) to body weight (SW) ratios.
# descending order to get the top three
top_three_families = family_brain_body_ratio.sort_values(ascending=False).head(3)

#top three families
top_three_families
```

```python
#i. Look up the Familiy with the highest brain to body weight ratio. Based on your experience, is this Family particularly smart?

highest_ratio_family = family_brain_body_ratio.idxmax()
highest_ratio_family
```

```python
# a.Read the dataset from the downloaded  from Kaggle to the csvfile.
import pandas as pd

df = pd.read_csv('batting/test.csv')

df.head()
```

```python
# b.Which player has the highest mean number of runs per year played? Your answer should be of type Series.



df.Runs = pd.to_numeric(df.Runs, errors='coerce')
df['StartYear'] = df['Span'].apply(lambda x: int(x.split('-')[0]))
df['EndYear'] = df['Span'].apply(lambda x: int(x.split('-')[1]))
df['YearsPlayed'] = df['EndYear'] - df['StartYear'] + 1
df['MeanRunsPerYear'] = df['Runs'] / df['YearsPlayed']
top_player = df.loc[df['MeanRunsPerYear'].idxmax()]
result = pd.Series({'Player': top_player['Player'], 'MeanRunsPerYear': top_player['MeanRunsPerYear']})
result
```

```python
# c.Provide a list of the top 3 players with the highest mean number of runs per year played? Your list should only include the columns for Player and the mean number of runs per year played and be of type DataFrame.
top_3_players = df[['Player', 'MeanRunsPerYear']].sort_values(by='MeanRunsPerYear', ascending=False)
top_3_players.head(3)
```

```python
# d.How many players are within 400 of the highest mean number of runs per year played? List the players and their mean number of runs as a DataFrame, and provide a count of the number of players using Python code (not just counting by eye/hand).

highest_mean = df['MeanRunsPerYear'].max()
players_within_400 = df[df['MeanRunsPerYear'] >= (highest_mean - 400)][['Player', 'MeanRunsPerYear']]
count_players_within_400 = players_within_400.shape[0]
count_players_within_400
```

```python
#e. Provide a list of the highest mean number of runs per year played by country.
df['Country'] = df['Player'].str.extract(r'\((.*?)\)')
mean_runs_by_country = df.loc[df.groupby('Country')['MeanRunsPerYear'].idxmax()][['Country', 'Player', 'MeanRunsPerYear']]
mean_runs_by_country.sort_values(by='MeanRunsPerYear', ascending=False)
```
